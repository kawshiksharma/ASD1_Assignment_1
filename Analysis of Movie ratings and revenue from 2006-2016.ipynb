```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
count_movies = df.shape[0]
count_movies
```

```python
df=df.sort_values(by='Rating', ascending=False)
```

```python
#filering out movies with les than 25,000 votes sicne IMDB doesn't count them as eligible for top movies

df_votes = (df["Votes"] > 100000)
df = df.loc[df_votes]
df.head(20)
count_movies = df.shape[0]
count_movies
```

```python
plt.figure(figsize=(12,10))
plt.scatter(df['Revenue (Millions)'], df['Rating'], s=10, c="green", cmap="green", edgecolor="black",alpha=0.75)

#plt.xscale("log")
#plt.yscale("log")

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Domestic Revenue", fontsize=14)
plt.ylabel("IMDB Ratings", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.scatter(df['Metascore'], df['Rating'], s=10, c="blue", cmap="green", edgecolor="black",alpha=0.75)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Metascore", fontsize=14)
plt.ylabel("IMDB Ratings", fontsize=14)
plt.show()
```

```python

plt.figure(figsize=(10,6))

plt.bar(df['Year'], df['Revenue (Millions)'],0.5, color="orange")
plt.tight_layout(pad=2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Domestic Revenue",fontsize = 14)
plt.title("Yearly Domestic Revenue Analysis",fontsize = 16, fontweight='bold')
plt.grid(True)
plt.show()
```
